## Baseline M4 Forecasting Demo

This notebook loads a subsample of M4 data and plots resulting forecasts of a baseline moving average model. 

```python
import sys
import pandas as pd
from pathlib import Path

# Add project root to Python path
project_root = Path.cwd().parent
sys.path.append(str(project_root))

# Import helper modules
from src.models.moving_average import MovingAverageLocalModel
from src.data.preprocessing import temporal_train_test_split
from src.evaluation.utils import plot_forecast
from src.data.preprocessing import plot_sampled_series
from src.data.load_m4 import build_processed_daily_table
```

```python
RAW = Path("../data/raw")
df = build_processed_daily_table(raw_data_dir=RAW, n_per_category=200)
```

```python
plot_sampled_series(df, 'Macro')
```

```python
train_df, test_df = temporal_train_test_split(df)
model = MovingAverageLocalModel(lags=(1, 7, 14, 21, 28))
model.fit(train_df)

predictions = model.predict(horizon=14)
```

```python
plot_forecast(train_df, test_df, predictions, 'D21')
```
